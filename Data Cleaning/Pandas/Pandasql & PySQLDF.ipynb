```python
#executing union all statements
q = """
        SELECT
            date
            , 'beef' AS meat_type
            , beef AS value
        FROM meat
        
        UNION ALL
        
        SELECT
            date
            , 'veal' AS meat_type
            , veal AS value
        FROM meat
        
        UNION ALL
        
        SELECT
            date
            , 'pork' AS meat_type
            , pork AS value
        FROM meat
        
        UNION ALL
        
        SELECT
            date
            , 'lamb_and_mutton' AS meat_type
            , lamb_and_mutton AS value
        FROM meat
        ORDER BY 1
    """
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
type(iris)
df = iris.data
```

```python
#IRIS is now in numpy array format
iris.data[0:5,]
```

```python
# convert IRIS to a data frame object
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
type(iris_df)
iris_df.head(5)
iris_df.columns
```

```python
# join the target variable to the iris df
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
iris_df.head()
```

```python
# replace the '(cm)' in the column header to NULL/blank to clean the feature names
print(iris_df.columns[0])
iris_df.columns = [re.sub("cm", "", col) for col in iris_df.columns]
```

```python
iris_df.columns
```

```python
# replace the '(cm)' in the column header to NULL/blank to clean the feature names
print(iris_df.columns[0])
iris_df.columns = [re.sub("[()]", "", col) for col in iris_df.columns]


# square bracket means include everything in bracket
# if you want to use particular column just mentioned column index iris_df.columns[0]
```

```python
iris_df.head()
```

```python
# print first 10 records from the iris df table
print(sqldf("SELECT * FROM iris_df LIMIT 10;"))
```

```python
# select sepal width column and species column and print first 10 records
print(sqldf("SELECT species FROM iris_df LIMIT 10;"))
```

```python
# let's compute some statistical functions from the table
# compute the mean, minimum and maximum values of sepal width feature and group by the species category
```

```python
# the query to perform the above task would look like as below
q = """
      select
        species
        , avg(sepalwidthcm)
        , min(sepalwidthcm)
        , max(sepalwidthcm)
      from
        iris_df
      group by
        species;
        
"""
```

```python
print(q)
```

```python
print(sqldf(q))
```

```python
#let's create a new table using inner join on the iris df by creating two dummy tables a and b and join by species column
q = """
    select
        a.*
    from
        iris_df a
    inner join
        iris_df b
            on a.species = b.species
    limit 10;
"""
```

```python
print(q)
```

```python
# use a filter criteria to select some records where the condition is met
```

```python
q = """
    select
        *
    from
        iris_df
    where
        species = 'virginica'
        and sepallengthcm > 7.7;
"""
```

```python
print(q)
```

```python
print(pysqldf(q))
```

```python
# create one unique id column in iris df
iris_df['id'] = range(len(iris_df))
```

```python
# fetch the records where a logical condition is met
q = """
    select
        *
    from
        iris_df
    where
        id in (select id from iris_df where sepalwidth*sepallength > 25);
"""
```
