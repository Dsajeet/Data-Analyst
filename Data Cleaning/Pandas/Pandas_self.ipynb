```python
# HTML Parameters
# 1.match : str or compiled regular expression, optional
# The set of tables containing text matching this regex or string will be returned. 

# 2.header : int or list-like or None, optional
# The row (or list of rows for a MultiIndex) to use to make the columns headers.

def get_forex_buy_quote(currency_code: str = 'EUR', source: str = 'FNB', order_type: str = 'sell'):
    """Get latest forex from FNB website

    """
    if source == 'FNB':
        tables = pd.read_html(
            'https://www.fnb.co.za/Controller?nav=rates.forex.list.ForexRatesList',
            index_col=1, header=0, match=currency_code)

        df = tables[0]

        types = {
            'buy': 'Bank Selling Rate',
            'sell': 'Bank Buying Notes',
        }

        exhange_rate = df.loc[currency_code, types[order_type]]
        print(df)
        return(float(exhange_rate)) 
    
get_forex_buy_quote() 
```
